```python
import numpy as np
import pandas as pd
```

```python
#사람 성과 인구 비율 파일 업로드
rawdata = pd.read_csv("/content/aaaa.txt",delimiter = "\t",header=None) 
rawdata

#rawdata에서 사람 성씨만 골라내기
surname = rawdata.iloc[0:-1,1]
for i in range(len(surname)):
  surname.iloc[i,] = surname.iloc[i,].split("(")[0]
surname

#우리나라 이름 종류
rawdata2 = pd.read_csv("/content/bbbb.txt",delimiter = "\t",header=None) 
middlename = rawdata2.iloc[0:-1,0]
middlename

#성씨 마다의 인구수
li = []
number = rawdata.iloc[0:-1,2]

#데이터에 있는 인구수는 숫자가 아니라 ,라는 문자가 포함된 string
#정규 표현식으로 제거
print(f"바꾸기 전 데이터 타입 : {number.iloc[0]} , {type(number.iloc[0])}")
import re
for i in range(len(number)):
  number[i] = int(re.sub(r'[^0-9]', '', number[i]))


#인구수 모두 합치기
sum = number.sum()
sum

#성씨마다의 인구수 비율
proba = pd.Series([])
for i in range(len(number)):
  proba[i] = number[i]/sum
proba

#만들 사람 수 반환
number = int(input("만들 사람 수 : "))

#성씨의 비율 별로 이름을 선정
name = []
sur=surname.sample(number,replace=True,weights=proba).reset_index(drop=True)
middle = middlename.sample(number,replace=True).reset_index(drop=True)
for i in range(len(sur)):
  name.append(sur[i]+middle[i])
print(f"만들어진 사람은 {len(name)}명 입니다.")

#주민번호 앞자리 무작위 만들기
import random

month_day = [31,29,31,30,31,30,31,31,30,31,30,31]
front = pd.Series([])
for i in range(number):
  year = (np.random.randint(1960,2023))
  month = (np.random.randint(1,12))
  day = (np.random.randint(1,month_day[month-1]))
  date = str(year) + "-" + str(month) + "-" + str(day)
  pd_date = pd.to_datetime(date).strftime('%Y%m%d')
  front[i] = pd_date[2:len(pd_date)]

#주민번호 뒷자리 만들기
import random
 
back = random.sample(range(1000000,4999999),number)

#주민번호 합치기
RNN = pd.Series([])
for i in range(number):
  RNN[i] = front[i] + "-" + str(back[i])

#주민번호로부터 age 만들기
#23이하부터는 20을 붙이고 나머지는 19를 붙이기
#현재년도에서 생년월일 빼고 생일이 안 지났으면 1을 빼준다.
from datetime import datetime

age = pd.Series([])

for i in range(len(RNN)):

  if int(RNN[i][0:2]) > 23:
    date = "19" + RNN[i][0:6]
    old = int(datetime.now().year) - int(date[0:4])
    if int(datetime.now().strftime('%m%d')) < int(RNN[i][2:6]):
      old -= 1
  else :
    date = "20" + RNN[i][0:6]
    old = int(datetime.now().year) - int(date[0:4])
    if int(datetime.now().strftime('%m%d')) < int(RNN[i][2:6]):
      old -= 1
  age[i] = old

#전화번호 만들기
bphone = random.sample(range(1000000000,1099999999),number)

phone = pd.Series([])

for i in range(len(bphone)):
  phone[i] = str(str(bphone[i]).zfill(11))

patient_info = pd.DataFrame({"pname":name,"pRNN":RNN,"phone":phone,"age":age})
patient_info

#파일로 저장할지 결정 y yes n no

conclude = input("저장?( 할거면 y 아니면 n ):")

if conclude == "y":
  path = input("저장할 경로 :")
  realpath = f"{path}/patient_info"
  patient_info.to_csv(realpath,encoding="utf-8-sig")

```
